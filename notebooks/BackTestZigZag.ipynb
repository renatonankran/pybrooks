```python
import talib as ta
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import sys
import os
sys.path.append(os.path.abspath('../main/lib'))
from zigzag import zig_zag
from bars import gap
```

```python
data = pd.read_csv("..\Datasets\AUDCADm_M30_2021.csv", sep='\t', 
names=['date', 't', 'open', 'high', 'low', 'close', 'vol1', 'vol2', 'vol3'],
parse_dates={'time':['date','t']})
columns = ['time', 'open', 'high', 'low', 'close']
# columns = ['time', 'open', 'high', 'low', 'close', 'zz', 'highs', 'lows']
data = data.loc[:, columns]
# data['time']=pd.to_datetime(data['time'], unit='ms')
data.tail()
```

```python
datam5 = pd.read_csv("..\Datasets\AUDCADm_M5_novembro_with_zz.csv")
columns = ['time', 'open', 'high', 'low', 'close']
# columns = ['time', 'open', 'high', 'low', 'close', 'zz', 'highs', 'lows']
datam5 = datam5.loc[:, columns]
# datam5['time']=pd.to_datetime(datam5['time'], unit='ms')
datam5.tail()
```

```python
zz, lows, highs, start = zig_zag(datam5,4)
```

```python
datam5=datam5.assign(start=start, zz=zz, highs=highs, lows=lows)
```

```python
datam5['kama17']=ta.KAMA(datam5['close'], timeperiod=17)
datam5.head(30)
```

```python
fig = go.Figure(data=[go.Candlestick(x=data['time'],
open=data['open'],
                high=data['high'],
                low=data['low'],
                close=data['close'])])

fig.add_trace(go.Scatter(
    x=data.loc[(data['zz'].notnull())]['time'],
    y=data.loc[(data['zz'].notnull())]['zz']
))
fig.add_trace(go.Scatter(
    x=data.loc[(data['start'].notnull())]['time'],
    y=data.loc[(data['start'].notnull())]['start'],
    mode = "markers",
    marker = dict(color = "black")
))


fig.update_xaxes(
    rangebreaks=[
        dict(bounds=["sat", "mon"]), #hide weekends
    ]
)
fig.update_layout(xaxis_rangeslider_visible=False)

fig.show()
```

```python
fig = go.Figure(data=[go.Candlestick(x=datam5['time'],
open=datam5['open'],
                high=datam5['high'],
                low=datam5['low'],
                close=datam5['close'])])

fig.add_trace(go.Scatter(
    x=datam5.loc[(datam5['zz'].notnull())]['time'],
    y=datam5.loc[(datam5['zz'].notnull())]['zz']
))
fig.add_trace(go.Scatter(
    x=datam5['time'],
    y=datam5['kama17']
))
fig.add_trace(go.Scatter(
    x=datam5.loc[(datam5['start'].notnull())]['time'],
    y=datam5.loc[(datam5['start'].notnull())]['start'],
    mode = "markers",
    marker = dict(color = "black")
))


fig.update_xaxes(
    rangebreaks=[
        dict(bounds=["sat", "mon"]), #hide weekends
    ]
)
fig.update_layout(xaxis_rangeslider_visible=False)

fig.show()
```
