```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import sys
import os
sys.path.append(os.path.abspath('../main/lib'))
from zigzag import zig_zag
from bars import gap
```

```python
columns = ['date','t', 'open', 'high', 'low', 'close','vol1','vol2','vol3']
data = pd.read_csv("..\Datasets\AUDCADm_M5_novembro_with_zz.csv")
columns = ['time', 'open', 'high', 'low', 'close', 'zz', 'highs', 'lows']
data = data.loc[:, columns]
# data['time']=pd.to_datetime(data['time'], unit='ms')
data.tail()
```

```python
gap(data.iloc[5],data.iloc[8])
```

```python
zz, lows, highs=zig_zag(data,4)
```

```python
data=data.assign(zz4=zz,lows4=lows,highs4=highs)
data.head()
```

```python
data.to_csv("..\Datasets\AUDCADm_M5_novembro_with_zz.csv")
```

```python
fig = go.Figure(data=[go.Candlestick(    open=data['open'],
                high=data['high'],
                low=data['low'],
                close=data['close'])])

fig.add_trace(go.Scatter(
    x=data.loc[(data['zz'].notnull())].index,
    y=data.loc[(data['zz'].notnull())]['zz']
))


fig.update_xaxes(
    rangebreaks=[
        dict(bounds=["sat", "mon"]), #hide weekends
    ]
)
fig.update_layout(xaxis_rangeslider_visible=False)

fig.show()
```
